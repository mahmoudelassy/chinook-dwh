```python
import os
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
from dotenv import load_dotenv

load_dotenv("dev.env")

USER = os.getenv("USER")
PASSWORD = os.getenv("PASSWORD")
HOST = os.getenv("HOST")
PORT = os.getenv("PORT")
DATABASE = os.getenv("DATABASE")

engine = create_engine(
    f"postgresql+psycopg2://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}"
)
```

```python
with engine.connect() as conn:
    print("Connected successfully!")
```

```python
inspector = inspect(engine)

tables = inspector.get_table_names(schema="public")

tables
```

```python
dfs = {}

for table in tables:
    dfs[table] = pd.read_sql_table(
        table,
        engine,
        schema="public"
    )

dfs.keys()
```

```python
from IPython.display import display, Markdown
def explore(df):
    display(Markdown("## Shape"))
    display(df.shape)

    display(Markdown("## Data Types"))
    display(df.dtypes)

    display(Markdown("## Missing Values"))
    display(df.isnull().sum())

    display(Markdown("## First 5 Rows"))
    display(df.head())

    display(Markdown("## Descriptive Statistics"))
    display(df.describe(include="all"))
```

# Artist Table

```python
artists=dfs['artist']
artists['artist_id']=artists['artist_id'].astype(str)
explore(artists)
```

# Album Table

```python
albums=dfs['album']
albums['artist_id']=albums['artist_id'].astype(str)
albums['album_id']=albums['album_id'].astype(str)
explore(albums)
```

# Employee Table

```python
employees=dfs['employee']
employees['employee_id']=employees['employee_id'].astype(str)
explore(employees)
```

# Invoice Table

```python
invoices=dfs['invoice']
invoices['invoice_id']=invoices['invoice_id'].astype(str)
invoices['customer_id']=invoices['customer_id'].astype(str)
explore(invoices)
```

# Invoice_line Table

```python
invoices_lines=dfs['invoice_line']
invoices_lines['invoice_line_id']=invoices_lines['invoice_line_id'].astype(str)
invoices_lines['invoice_id']=invoices_lines['invoice_id'].astype(str)
invoices_lines['track_id']=invoices_lines['track_id'].astype(str)
explore(invoices_lines)
```

# Track Table

```python
tracks=dfs['track']
tracks['track_id']=tracks['track_id'].astype(str)
tracks['album_id']=tracks['album_id'].astype(str)
tracks['media_type_id']=tracks['media_type_id'].astype(str)
tracks['genre_id']=tracks['genre_id'].astype(str)
explore(tracks)
```
